```python
# ============================================
# ONE-CLICK UPLOAD SCRIPT
# Copy ALL of this into ONE Colab cell and run
# ============================================

import os
import shutil
from IPython.display import display, HTML

# Create folders
os.makedirs('test_images', exist_ok=True)

# Show upload instructions
display(HTML("""
<style>
.upload-box {
    padding: 20px;
    background: #f0f8ff;
    border: 2px dashed #2E8B57;
    border-radius: 10px;
    text-align: center;
    margin: 20px 0;
}
.upload-title {
    color: #2E8B57;
    font-size: 1.5em;
    margin-bottom: 10px;
}
.upload-steps {
    text-align: left;
    display: inline-block;
    margin: 10px;
}
</style>

<div class="upload-box">
    <div class="upload-title">ğŸ“¤ UPLOAD YOUR IMAGES</div>
    <div class="upload-steps">
        <strong>1. Click the ğŸ“ folder icon on LEFT</strong><br>
        <strong>2. Click "Upload" button</strong><br>
        <strong>3. Select these 6 files:</strong><br>
        &nbsp;&nbsp;â€¢ cardboard_box.jpg<br>
        &nbsp;&nbsp;â€¢ glass_bottle.jpg<br>
        &nbsp;&nbsp;â€¢ metal_can.jpg<br>
        &nbsp;&nbsp;â€¢ paper_cup.jpg<br>
        &nbsp;&nbsp;â€¢ plastic_bottle.jpg<br>
        &nbsp;&nbsp;â€¢ trash_bag.jpg<br>
        <strong>4. Wait for upload to complete</strong><br>
        <strong>5. Come back here and continue</strong>
    </div>
</div>
"""))

# Wait for user to upload
input("â¸ï¸ Press ENTER after uploading all 6 images...")

print("\nğŸ”„ Processing uploaded files...")

# List of expected images
expected_images = [
    'cardboard_box.jpg', 'glass_bottle.jpg', 'metal_can.jpg',
    'paper_cup.jpg', 'plastic_bottle.jpg', 'trash_bag.jpg'
]

# Check and move each image
print("\nğŸ“‹ Checking uploaded files:")
for img in expected_images:
    if os.path.exists(img):
        shutil.move(img, f'test_images/{img}')
        print(f"âœ… {img} â†’ test_images/")
    else:
        # Check with different cases
        found = False
        for file in os.listdir('.'):
            if img.lower() in file.lower() and file.lower().endswith(('.jpg', '.jpeg', '.png')):
                shutil.move(file, f'test_images/{img}')
                print(f"âœ… {file} â†’ test_images/{img}")
                found = True
                break
        if not found:
            print(f"âŒ {img} - NOT FOUND")

# Show results
print("\n" + "="*50)
print("ğŸ“Š UPLOAD RESULTS:")
print("="*50)

if os.path.exists('test_images'):
    files_in_folder = os.listdir('test_images')
    print(f"ğŸ“ test_images/ contains {len(files_in_folder)} files:")

    for file in sorted(files_in_folder):
        file_path = os.path.join('test_images', file)
        size_kb = os.path.getsize(file_path) / 1024
        print(f"   ğŸ–¼ï¸ {file} ({size_kb:.1f} KB)")

    if len(files_in_folder) == 6:
        print("\nğŸ‰ SUCCESS! All 6 images uploaded!")
        print("âœ… Ready to use with your model")
    else:
        print(f"\nâš ï¸ Only {len(files_in_folder)}/6 images found")
        print("Check if all files uploaded correctly")
else:
    print("âŒ test_images/ folder not created")

print("\nğŸš€ Next: Upload your 'waste_classifier.h5' model file")
```
